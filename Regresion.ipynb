## ¿Qué es Regresión?

La **regresión** es una técnica estadística y de aprendizaje automático utilizada para **modelar la relación entre una variable dependiente (objetivo) y una o más variables independientes (predictoras)**.

El objetivo principal de la regresión es **predecir valores numéricos continuos** a partir de datos observados.

En términos simples, la regresión busca responder preguntas como:

- ¿Cómo cambia una variable cuando otra variable cambia?
- ¿Podemos predecir un valor futuro a partir de datos históricos?

Por ejemplo:


| Variable independiente | Variable dependiente | Objetivo |
|-----------------------|---------------------|-------------------------------|
| Tamaño de una casa | Precio de la casa | Predecir el precio de la casa en función de su tamaño |
| Horas de estudio | Nota obtenida | Estimar la calificación esperada según las horas de estudio |
| Temperatura | Consumo de energía | Predecir el consumo energético en función de la temperatura |

En todos estos casos, queremos encontrar una **relación matemática** que nos permita estimar el valor de la variable dependiente.

## Intuición de la regresión

La regresión intenta encontrar una **función matemática** que describa la relación entre variables.

En el caso más simple, la **regresión lineal**, la relación entre variables se representa como una línea recta. Esto puede hipotetizarse a partir de scatter plots bivariados.

Para profundizar en la intuición y el funcionamiento de la **regresión lineal**, se recomienda ver los siguientes videos antes de continuar con la lectura del Notebook:

- https://www.youtube.com/watch?v=k964_uNn3l0  
- https://www.youtube.com/watch?v=HXahn-XGn2Q

La ecuación básica para una regresion lineal simple es:

$$
y = \beta_0 + \beta_1 x
$$

donde:

- $y$ : variable dependiente (lo que queremos predecir)
- $x$ : variable independiente
- $\beta_0$ : intercepto
- $\beta_1$ : pendiente

El objetivo del modelo es **encontrar los valores de los parámetros** que mejor ajusten la relación entre los datos.

### Paso 3: Entrenamiento del modelo de regresion
En este paso se entrenará el **modelo de regresión** para estimar el **tiempo de servicio de las máquinas dispensadoras**.

Dada la relación lineal observada en el paso 2, y considerando que el tiempo de suministro puede ser predicho por dos variables predictoras, se entrenará un modelo de **regresión lineal múltiple**. El modelo utilizará como variables predictoras:

- **envases**: número de envases que deben ser reemplazados
- **distancia**: distancia que debe recorrer el técnico hasta la máquina

La variable que queremos predecir es:

- **tiemposum**: tiempo total requerido para realizar el servicio.

Para entrenar el modelo utilizaremos la librería **Scikit-learn**, una de las herramientas más utilizadas en **Machine Learning en Python**.

Ejecute la siguiente celda para entrenar el modelo.

```python
from sklearn.linear_model import LinearRegression

# Variables predictoras
X = df[['envases', 'distancia']]

# Variable objetivo
y = df['tiemposum']

# Crear modelo
modelo = LinearRegression()

# Entrenar modelo
modelo.fit(X, y)

print("Modelo entrenado correctamente")
```

```python
## Visualizar parametros del modelo
print("Intercepto:", modelo.intercept_)
print("Coeficientes:", modelo.coef_)
```

### **Paso 6: Análisis de supuestos del modelo**

Después de entrenar un modelo de **regresión lineal múltiple**, es importante revisar algunos supuestos estadísticos. Estos supuestos permiten validar si la regresión lineal es adecuada para describir la relación entre las variables.

En esta sección se evaluarán principalmente los supuestos mas comunes:

- **Normalidad de los residuos**
- **Homocedasticidad**
- **Independencia de los errores**

El cumplimiento de estos supuestos ayuda a garantizar que las inferencias y predicciones del modelo sean confiables.

##### **Prueba de normalidad de los residuos**

Uno de los supuestos del modelo de **regresión lineal** es que los **residuos siguen aproximadamente una distribución normal**.

Para evaluar este supuesto se utiliza la **prueba de Shapiro–Wilk**, una de las pruebas estadísticas más comunes para verificar normalidad.

**Hipótesis:**

- **H₀:** los residuos siguen una distribución normal
- **H₁:** los residuos no siguen una distribución normal

**Criterio de decisión:**

- Si **p-value > 0.05**, no se rechaza \(H_0\) y se puede asumir normalidad.
- Si **p-value < 0.05**, se rechaza \(H_0\), lo que sugiere que los residuos no son normales.

```python
from scipy.stats import shapiro



# Calcular residuos
residuos = y - pred
# prueba de normalidad
stat, p_value = shapiro(residuos)

print("Prueba de normalidad Shapiro-Wilk")
print("----------------------------------")
print(f"Estadístico: {stat:.4f}")
print(f"p-value: {p_value:.4f}")
```

**Conclusión de la prueba de normalidad**

El **p-value obtenido (0.3321)** es mayor que 0.05, por lo que **no se rechaza la hipótesis nula de normalidad**.

Esto indica que **los residuos del modelo pueden considerarse aproximadamente normales**, por lo que el supuesto de normalidad del modelo de regresión se cumple.

##### **Prueba de homocedasticidad de los residuos**

La **homocedasticidad** significa que la varianza de los residuos se mantiene aproximadamente constante a lo largo de los valores predichos.

En otras palabras, el error del modelo no debería aumentar ni disminuir de forma sistemática.

Una de las pruebas más utilizadas para evaluar este supuesto es la **prueba de Breusch–Pagan**.

**Hipótesis:**

- **H₀:** los residuos tienen varianza constante (**homocedasticidad**)
- **H₁:** los residuos no tienen varianza constante (**heterocedasticidad**)

**Criterio de decisión:**

- Si el **p-value > 0.05**, no se rechaza \(H_0\), por lo que se asume homocedasticidad.
- Si el **p-value < 0.05**, se rechaza \(H_0\), lo que sugiere heterocedasticidad.

```python
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan

# Residuos del modelo
residuos = y - pred

# Agregar constante a X para la prueba
X_const = sm.add_constant(X)

# Prueba de Breusch-Pagan
bp_test = het_breuschpagan(residuos, X_const)

bp_labels = [
    "Lagrange multiplier statistic",
    "p-value",
    "f-value",
    "f p-value"
]

print("Prueba de Breusch-Pagan")
print("------------------------")
for label, value in zip(bp_labels, bp_test):
    print(f"{label}: {value:.4f}")
```

**Conclusión de la prueba de Breusch–Pagan**

El **p-value obtenido (0.7174)** es mayor que 0.05, por lo que **no se rechaza la hipótesis nula de homocedasticidad**.

Esto indica que **no existe evidencia de heterocedasticidad en los residuos**, lo que sugiere que la varianza de los errores del modelo puede considerarse aproximadamente constante.

Por lo tanto, **el supuesto de homocedasticidad del modelo de regresión se cumple** para este conjunto de datos.

### **Prueba de independencia de los residuos**

Otro supuesto importante en regresión lineal es la **independencia de los errores**. Esto significa que los residuos no deben estar correlacionados entre sí.

Este supuesto es especialmente importante en datos de tipo temporal o secuencial, aunque también puede revisarse en otros contextos.

Una de las medidas más utilizadas para evaluar este supuesto es el **estadístico de Durbin–Watson**.

**Interpretación general del estadístico Durbin–Watson:**

- **≈ 2**: no hay autocorrelación
- **< 2**: posible autocorrelación positiva
- **> 2**: posible autocorrelación negativa
- **≈ 0**: fuerte autocorrelación positiva
- **≈ 4**: fuerte autocorrelación negativa

```python
from statsmodels.stats.stattools import durbin_watson

dw = durbin_watson(residuos)

print("Prueba de independencia de errores")
print("----------------------------------")
print(f"Durbin-Watson: {dw:.4f}")
```

**Conclusión de la prueba de Durbin–Watson**

El estadístico **Durbin–Watson obtenido (2.1312)** es cercano a **2**, lo que indica que **no existe evidencia de autocorrelación en los residuos**.

Por lo tanto, se puede considerar que **el supuesto de independencia de los errores se cumple en el modelo**.

### Paso 3: Entrenamiento del modelo de regresion

Como se ha mencionado anteriormente, además de la regresión lineal múltiple, es posible utilizar modelos más **flexibles** que permitan capturar relaciones más complejas entre las variables.

En este caso, entrenaremos un modelo alternativo que no se limita a relaciones estrictamente lineales, lo que puede resultar útil cuando existen **patrones no lineales** en los datos.

Existen diversas opciones de modelos de regresión que pueden emplearse, entre ellas:

- **Regresión lineal múltiple** (modelo base) (SI requiere validacion de supuestos)
- **Árboles de decisión** (No requiere validacion de supuestos)
- **Random Forest** (No requiere validacion de supuestos)
- **Gradient Boosting (por ejemplo, XGBoost)** (No requiere validacion de supuestos)

Cada uno de estos modelos tiene ventajas y limitaciones. Mientras que la regresión lineal es fácil de interpretar, modelos como los árboles de decisión pueden adaptarse mejor a estructuras complejas en los datos.

En este paso, el objetivo es **entrenar un modelo de regresión alternativo** y analizar su capacidad para predecir la variable objetivo.

Posteriormente, se podrán comparar los resultados obtenidos con diferentes modelos para determinar cuál ofrece un mejor desempeño en este problema.

```python
# Definir la variable objetivo (target)
# Esta es la variable que queremos predecir: la probabilidad de admisión
y = df['Chance of Admit ']

# Definir las variables predictoras (features)
# Se eliminan:
# - 'Serial No' porque es solo un identificador y no aporta información al modelo
# - 'Chance of Admit ' porque es la variable objetivo
X = df.drop(['Serial No', 'Chance of Admit '], axis=1)
```

#### Paso 3.1: División de datos en entrenamiento y prueba (Train/Test Split)

En este paso se divide el dataset en dos subconjuntos:

- **Datos de entrenamiento (train):** utilizados para entrenar el modelo.
- **Datos de prueba (test):** utilizados para evaluar el desempeño del modelo en datos no vistos.

Esta división es una **buena práctica en Machine Learning**, ya que permite medir qué tan bien el modelo generaliza a nuevos datos y ovitar overfitting. Se recomienda que el estudiante se familiarice con el concepto de overfitting. algunos materiales de referencia incluyen:

**¿Por qué es importante?**

Si evaluamos el modelo utilizando los mismos datos con los que fue entrenado, podemos obtener resultados demasiado optimistas. Esto ocurre porque el modelo puede "memorizar" los datos en lugar de aprender patrones generales, lo que se conoce como **sobreajuste (overfitting)**.

Al separar los datos, podemos simular un escenario más realista donde el modelo debe hacer predicciones sobre información nueva.

### Implementación

Se utiliza la función `train_test_split` de `scikit-learn`, donde:

- `train_size=0.8` indica que el 80% de los datos se usa para entrenamiento
- `random_state=2529` es una semila que asegura que la división sea reproducible. puedes seleccionar cualquier numero.

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
from sklearn.model_selection import train_test_split

# X: variables predictoras
# y: variable objetivo

# Se divide el dataset en:
# - 80% para entrenamiento (train)
# - 20% para prueba (test)
# random_state garantiza que la división sea reproducible

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    train_size=0.8,
    random_state=2529
)
```

```python
# Importar el modelo de árbol de decisión para regresión y la función para visualizar el árbol
from sklearn.tree import DecisionTreeRegressor, plot_tree

# Crear el modelo de árbol de decisión
# max_depth=6 especifica la profundidad del árbol
# random_state=2529 asegura que los resultados sean reproducibles
dtr = DecisionTreeRegressor(max_depth=6, random_state=2529) ##Notese que aqui se selecciona el modelo a utilizar, en este caso arboles de decision. sin embargo se podria seleccionar cualqueir modelo de sklearn, incluyendo regresion lineal multiple vista en el caso 1. (se recomienda que el estudiante explore los modelos de regresion contenidos en https://scikit-learn.org/stable/supervised_learning.html)
```

#### Validación cruzada del modelo (Cross-Validation)

Después de separar el conjunto de datos en entrenamiento y prueba, una practica comun es entrenar el modelo utilizando validacion cruzada sobre los datos de entrenamiento. Se recomienda que el estudiante se familiarice con el termino validacion cruzada. algunos recursos relevantes incluyen:

https://machinelearningmastery.com/k-fold-cross-validation/

https://www.youtube.com/watch?v=v6a5bGVhPE0

https://deepnote.com/app/a_mas/Cross-Validation-en-Python-685fa851-b5b2-4c5b-b5fb-3dc5ae64838f

A diferencia de una única división entre entrenamiento y prueba, la validación cruzada consiste en dividir el conjunto de entrenamiento en múltiples subconjuntos (folds). En cada iteración:

- El modelo se entrena con una parte de los datos
- Se valida con la parte restante
- Este proceso se repite varias veces (en este caso, **5 folds**)

**¿Por qué usar cross-validation?**

- Reduce la dependencia de una sola partición de los datos
- Proporciona una estimación más estable del desempeño del modelo agregando datos de validacion y despues probando el modelo ajustado con datos de prueba.
- Permite detectar variabilidad en los resultados
- Ayuda a identificar posibles problemas de overfitting

En este caso, vamos a utilizar `cv=5`, lo que significa que el conjunto de entrenamiento se divide en **5 particiones**, y el modelo se evalúa 5 veces utilizando diferentes combinaciones de entrenamiento y validación.

```python
# Validar el modelo utilizando validación cruzada (cross-validation)
from sklearn.model_selection import cross_validate

# Se evalúa el modelo dtr utilizando únicamente los datos de entrenamiento
# cv=5 indica que se realizará una validación cruzada con 5 particiones (folds)
# El modelo se entrena y valida 5 veces con diferentes subconjuntos de los datos
# Esto permite obtener una estimación más robusta del desempeño del modelo

cross_validate(dtr, X_train, y_train, cv=5)
```

La validación cruzada genera múltiples valores de desempeño (R²), uno por cada partición del conjunto de datos.

En este caso, los valores obtenidos fueron:

$$
[0.42,\ 0.59,\ 0.50,\ 0.77,\ 0.67]
$$

El promedio de estos valores es aproximadamente **0.59**, lo que indica el desempeño (R²) general del modelo. valores mayores a 0.7 son generalmente aceptados en la practica aunque al entrenar varios modelos se elige el que proorcione el valor mas alto. En este caso, 0.59 es un valor relativamente bajo que demuestra un ajuste deficiente del modelo a los datos.

Se observa cierta variabilidad entre los resultados, lo que sugiere que el modelo puede comportarse de manera diferente dependiendo de los datos utilizados en el entrenamiento.

En general, la validación cruzada proporciona una estimación más confiable del desempeño del modelo en comparación con una única evaluación.

### Ajuste de hiperparámetros con Grid Search

Una vez entrenado el modelo base de **árbol de decisión**, es posible mejorar su desempeño ajustando algunos de sus **hiperparámetros**.

Los hiperparámetros son configuraciones del modelo que se definen antes del entrenamiento y que influyen directamente en su comportamiento. En el caso de un árbol de decisión, algunos de los más importantes son:

- **max_depth**: controla la profundidad máxima del árbol (este lo colocamos por defecto como 6 arriba)
- **criterion**: define la función utilizada para medir la calidad de una división

En este paso se utilizará la técnica **Grid Search con validación cruzada** (**GridSearchCV**), que permite:

- probar múltiples combinaciones de hiperparámetros
- evaluar cada combinación mediante validación cruzada
- seleccionar automáticamente la configuración con mejor desempeño

Este procedimiento es una buena práctica en Machine Learning, ya que ayuda a encontrar un modelo más adecuado y reduce la posibilidad de sobreajuste o de un ajuste deficiente (Underfitting).

```python
# Mostrar los hiperparámetros actuales del modelo de árbol de decisión
# Esto permite conocer la configuración inicial del modelo antes de optimizarlo
dtr.get_params()
```

```python
# Definir las combinaciones de hiperparámetros que se desean evaluar
# model__ indica que estos parámetros pertenecen al modelo dentro del Pipeline

params = {
    'model__max_depth': [2, 3, 4, 5, 6, 7, 8, 9, 10],   # Profundidad máxima del árbol
    'model__criterion': ['squared_error', 'absolute_error'],  # Función de error para dividir nodos
    'model__random_state': [2529]   # Semilla para asegurar reproducibilidad
}
```

```python
# Importar la clase Pipeline
# Pipeline permite organizar el flujo de trabajo del modelo dentro de una sola estructura
from sklearn.pipeline import Pipeline
```

```python
# Definir un pipeline que contiene el modelo de árbol de decisión
# En este caso solo se incluye un paso llamado 'model'
pipe = Pipeline([('model', dtr)])
```

```python
# Importar GridSearchCV
# Esta herramienta prueba múltiples combinaciones de hiperparámetros
# y selecciona la mejor utilizando validación cruzada
from sklearn.model_selection import GridSearchCV

# Crear el objeto GridSearchCV
# cv=5 indica que se usará validación cruzada con 5 particiones
grid = GridSearchCV(pipe, params, cv=5)
```

```python
# Mostrar la configuración del Grid Search
# Esto permite verificar el estimador, el pipeline y la grilla de parámetros
grid
```

```python
# Entrenar GridSearchCV con los datos de entrenamiento
# El algoritmo probará todas las combinaciones de hiperparámetros definidas
# y evaluará cada una utilizando validación cruzada
grid.fit(X_train, y_train)
```

```python
# Mostrar la mejor combinación de hiperparámetros encontrada
# según el desempeño obtenido durante la validación cruzada
grid.best_params_
```

El proceso de **Grid Search con validación cruzada** encontró la siguiente combinación óptima de hiperparámetros para el modelo de árbol de decisión:

- **criterion: squared_error**
- **max_depth: 3**

**Interpretación:**

- **criterion = squared_error**  
  Indica que el modelo utiliza el **error cuadrático medio (MSE)** como criterio para realizar las divisiones en el árbol. Este criterio busca minimizar la variabilidad dentro de cada nodo, lo que es adecuado para problemas de regresión.

- **max_depth = 3**  
  El árbol debe tener una profundidad máxima de 3 niveles. Esto sugiere que un modelo relativamente **simple** es suficiente para capturar los patrones en los datos. Además, limitar la profundidad ayuda a **evitar el sobreajuste (overfitting)**, mejorando la capacidad de generalización.

En conclusión, el modelo óptimo encontrado es un **árbol de decisión poco profundo**. El siguiente paso es reentrenar el modelo con los mejores hiperparametros encontrados y probarlo con los datos de prueba y ver su rendimiento antes de usarlo para predecir.

```python
# ================================
# 1. Seleccionar mejor modelo
# ================================
best = grid.best_estimator_

# ================================
# 2. Entrenar mejor modelo
# ================================
best.fit(X_train, y_train)

print("Modelo óptimo entrenado")

# ================================
# 3. Evaluación en entrenamiento
# ================================
train_score = best.score(X_train, y_train)
print(f"R² en entrenamiento: {train_score:.4f}")

# ================================
# 4. Predicciones en test
# ================================
y_pred = best.predict(X_test)

# ================================
# 5. Métricas de evaluación
# ================================
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\nEvaluación en conjunto de prueba")
print("--------------------------------")
print(f"MAE  (Error absoluto medio): {mae:.4f}")
print(f"MAPE (Error porcentual medio): {mape:.4f}")
print(f"R²   (Coeficiente de determinación): {r2:.4f}")

# ================================
# 6. Visualización del árbol
# ================================
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree, DecisionTreeRegressor

# Crear modelo final (mismos hiperparámetros óptimos)
final = DecisionTreeRegressor(max_depth=3, random_state=2529)
final.fit(X_train, y_train)

# Graficar árbol
fig, ax = plt.subplots(figsize=(15, 10))
plot_tree(final, feature_names=X.columns, filled=True)

plt.title("Árbol de decisión optimizado")
plt.show()
```
